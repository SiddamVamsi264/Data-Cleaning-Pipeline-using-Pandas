**Initial Data Validation and Profiling Checklist:**

Assess and impute or remove missing values

Identify and appropriately encode categorical variables

Validate and correct data type mismatches

Detect high cardinality in string/object-type columns

Investigate and flag outliers, zero, or negative values in numerical features

```python
missing = df.isnull().sum()
missing = missing[missing > 0].sort_values(ascending=False)
print(missing)
```

Lets deal with missing values

```python
cat_cols = df.select_dtypes(include='object').columns
num_cols = df.select_dtypes(include='number').columns

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())
```

Fixing data types

```python
df['LoanAmount']=df['LoanAmount'].astype(float)

for col in cat_cols:
    df[col] = df[col].astype('category')
df.head()
```

Lets standardize text columns


```python
for col in cat_cols:
    df[col] = df[col].str.strip().str.title()
df.head()
```

Now lets handle outliers

```python
for col in ['LoanAmount', 'ApplicantIncome']:
  upper = df[col].quantile(0.99)
  df[col]=df[col].apply(lambda x: upper if x>upper else x)
```

Now lets store this in another data file

```python
df.to_csv('cleaned_data.csv', index=False)
```

Let's build a pipeline to this

```python
def clean_loan_data(df):
    # fill missing values
    for col in df.select_dtypes(include='object'):
        df[col].fillna(df[col].mode()[0])
    for col in df.select_dtypes(include=['int64', 'float64']):
        df[col].fillna(df[col].median())

    # type conversions
    for col in df.select_dtypes(include='object'):
        df[col] = df[col].astype('category')
        df[col] = df[col].str.strip().str.lower()

    # outlier capping
    for col in ['LoanAmount', 'ApplicantIncome']:
        if col in df.columns:
            upper = df[col].quantile(0.99)
            df[col] = df[col].apply(lambda x: upper if x > upper else x)

    return df
```

Here is how we use this function

```python
df = pd.read_csv('/content/sample_data/Loan Prediction.csv')
cleaned_df = clean_loan_data(df)
cleaned_df.head()
```
